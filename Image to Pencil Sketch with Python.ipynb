```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('dog.jpg')
cv2_imshow(img)
```

```python
greyscale_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(greyscale_img)
```

```python
inverted_image = 255 - greyscale_img
cv2_imshow(inverted_image)

```

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred_image)
```

```python
inv_blur_img = 255 - blurred_image
cv2_imshow(inv_blur_img)
```

```python
sketch = cv2.divide(greyscale_img, inv_blur_img, scale=256.0)
cv2_imshow(sketch)
```

```python
cv2_imshow(img)
cv2_imshow(sketch)

```
